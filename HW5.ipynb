### Images in Python

An image is essentially a big array with 3 values that specify color intensities in the red, green and blue filter (RGB). Each coordinate pair in this array specifies a pixel on the picture. In Python, you can use a numpy array to store an image. Each element in this array should contain a number between 0 and 255. When we take one pixel (so one point in the first two dimensions, but all three colors), `[255, 255, 255]` represents a white pixel. `[0,0,0]` represents a black pixel and `[255,0,0]`, `[0,255,0]` and `[0,0,255]` represent red, green and blue respectively.

When an image is big, the resolution can be adjusted by undersampling your image. Undersampling means that instead of working with the full image, we only use every N'th column and row as our image. In this excercise we will undersample the 'Academy_building.jpg' image by a factor of N (N'th pixel!) in both the x and y directions after having loaded the entire image. ### Excercise 3: Grayscale

For many applications (simple models or just art) it can be useful to convert the original image to grayscale. There are several ways to convert a color image to grayscale, for this excercise we will set the grayscale value of a pixel as the mean value of the R,G and B values of said pixel. Output a two dimensional numpy array with the mean values over the third dimension.

```python
#grayscale
# in: a three dimensional numpy array: image
# out: a two dimensional numpy array containing the mean over the third dimension (or grayscale image)

def grayscale(image):
    mean_list = list()
    for row in image:
        row_mean = row.mean(axis=1)
        mean_list.append(row_mean)
        
    mean_array = np.vstack(mean_list)
    print(mean_array.shape)
        
        
    #Complete your function here
    return (mean_array)
```

### Excercise 4: Pixel Value Distribution
Next we will take a look at the distribution of the pixel values in this image. For this we will again be using the grayscale version of 'Academy_building.jpg'. To explore which values the pixels have, you can plot a histogram (you can try this yourselves). This histogram would show you a sort of bimodal distribution, with a steep normal curve above 175 and a more flat normal curve below 175.

We would like to summarize the upper normal curve (of all pixel values above 175) by computing the mean and the standard deviation. In order to select only the pixels with a value above 175 you can use a mask. A mask is a very convenient object that returns numpy array values based on a condition set by the user. For example, if we want to return all even values of an array the syntax could look as follows:

`even = array[array%2 == 0]`

```python
#summary_pixels
# in: two dimensional numpy array: grayscale_image
# out: tuple containing the mean and standard deviation of the pixel distribution over 175
def summary_pixels(grayscale_image):
    upper = grayscale_image[grayscale_image > 175]
    upper_mean = np.mean(upper)
    upper_sd = np.sqrt(np.var(upper))
    #print(upper_mean)
    #print(upper_sd)
    result = (upper_mean,upper_sd)
    #Complete your function here
    return (result)
```

### Excercise 5: Tiling

In this excercise we will tile a bigger array with the grayscale image of the Academy Building. Tiling is the process of filling an array with the same scalars or subarray entirely by repeating the values you put in. An example of an array tiled with the scalar 3 is given below:

$$\pmatrix{3&3&3&3\\3&3&3&3\\3&3&3&3}$$

In our case we will tile the image with the grayscaled Academy Building image and a transposed version of this image in a checkers board pattern. If we let $A$ be the original grayscale image and $T$ the transposed version, this should look something like this:

$$\pmatrix{A&T&A&T\\T&A&T&A\\A&T&A&T}$$

You are going to write a function that can output a tiled numpy array (or, grayscale image) given the number of tiles in rows and columns. Make sure that your code can also work with different (square) images.

```python
#tiling
# in: two-dimensional numpy array: grayscale_image
#     integer with number of columns of tiles: x_tiles
#     integer with number of rows of tiles: y_tiles
# out: two dimensional numpy array with x_tiles by y_tiles tiles with transpose checkered pattern

def tiling(grayscale_image, x_tiles, y_tiles):
    transpose = grayscale_image.T
   
    first = np.r_[[grayscale_image,transpose] * int(y_tiles/2) + [grayscale_image] * (y_tiles%2)]
    first_row = np.hstack(first)
    second = np.r_[[transpose,grayscale_image] * int(y_tiles/2) + [transpose] * (y_tiles%2)]
    second_row = np.hstack(second)
    #print(first_row)
    #print(second_row)
    checkboard = np.r_[[first_row,second_row] * int(x_tiles/2) + [first_row] * (x_tiles%2)]
    tiling = np.vstack(checkboard)
    
    #print(tiling)
    #Complete your function here
    return (tiling)
```
